```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
import datetime
from numpy import linalg
#from numpy import *
import pandas as pd
import os
%matplotlib inline
```

```python
N = 20
C = [0] * N
Ns = np.linspace(2,N+1, N)
epsilons = np.linspace(0.1,0.5,9)

fig, ax = plt.subplots(figsize=(20, 10))
legend = []
count = 0
for epsilon in epsilons:
    C[0] = 0.5+epsilon
    C[1] = 0.5+epsilon/2
    for i in range(N-2):
        C[i+2] = epsilon/(2+epsilon/C[i])+epsilon
    ax.plot(Ns, C)
    ax.set_xlabel('Number of Turns',fontsize=20)
    ax.set_ylabel(r'Equivlent winding capacitance (* $C_{turn \quad to \quad core}$)',fontsize=20)
    legend.append('epsilon = '+ str(float("{:.2f}".format(epsilons[count]))))
    count+= 1
ax.legend(legend,fontsize=20)
```
