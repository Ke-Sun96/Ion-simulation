```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn


plt.rc('font', size=20)
plt.rc('axes', titlesize=20)
plt.rc('axes', labelsize=20)
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rc('legend', fontsize=15)
plt.rc('figure', titlesize=20)

# Define the range of x values
x = np.linspace(0, 10, 400)

# Define the orders of the Bessel functions
orders = [0, 1, 2, 3, 4, 5]

# Plot Bessel functions for each order
plt.figure(figsize = (8,4))
for n in orders:
    plt.plot(x, jn(n, x), label=f'J{n}(x)')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('Jn(x)')
plt.title('Bessel Functions')
plt.legend(ncol = 3)
plt.grid(True)

# Show plot
# plt.show()
plt.savefig('Fig3_Bessel.png', dpi=300, bbox_inches='tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of L values
L = np.linspace(0.01, 1, 100)  # Starting from 0.01 to avoid division by zero

# Calculate the function values
y = np.pi * np.sqrt(1 - L) / L

# Plot the curve
plt.plot(L, y)
plt.xlabel('L')
plt.ylabel('$\pi \cdot \sqrt{(1-L)}/L$')
plt.title('Plot of $\pi \cdot \sqrt{(1-L)}/L$')
plt.grid(True)
plt.show()
```
