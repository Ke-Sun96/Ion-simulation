```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def waist(z, w0, wavelength = 0.369, n = 1):
    ## The length unit is um
    zR = np.pi*w0**2*n/wavelength
    return w0*np.sqrt(1+(z/zR)**2)
```

# Beam clipping calculation
This code calculates the optimal beam size required to prevent clipping at the trap's edge. Assuming the beam waist is centered within the trap, the beam will expand while it propagates according to the wave propagation function defined as
$$
\omega_z = \omega_0\sqrt{1+(\frac{z-z_0}{z_R})^2}
$$

Here we define the clearance as the ratio of the beam-surface distance $d \, (=74 \mu m$ for peregrine surface trap) and the beam size $(z_R)$ at the edge of the trap: $c = d/z_R$
## 355nm laser
### $90^{\circ}$ beam

```python
waists = np.linspace(0.01, 30, 101)
clearance = []
for w in waists:
    wz = waist(600, w, wavelength = 0.355)
    clearance.append(74/wz) # 74 is the distance between the ion and trap surface
plt.plot(waists, clearance)
plt.axvline(x = 5)
plt.xlabel('Waist (um)')
plt.ylabel('Clearance (74um/waist(z))')
max_index_col = np.argmax(clearance)
print('The max clearance is at: '+ str(waists[max_index_col])+'um')
```
