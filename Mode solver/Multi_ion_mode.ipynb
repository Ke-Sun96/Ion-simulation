```python
SMALL_SIZE = 16
MEDIUM_SIZE = 20
BIGGER_SIZE = 24

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

num_ions = np.linspace(2,50,49)
com_radial = 1.988e6
com_axial = np.sqrt(0.09*0.95)*1e6

mode_freq_exp = [1.728,1.789,1.845,1.889,1.929,1.959,1.988]
num_ions = [4,5,6,7]
plt.figure(figsize = (12,8))
for n_ions in num_ions:
    ionchain = ms.TrapModel(n_ions, com_radial, com_axial, 'cubic', q3 = 0.000, source='model', real_units=True)
    ionchain.calc_ion_pos()
    modes = ionchain.calc_modes()
    modes_spacing = [modes[i+1]-modes[i] for i in range(len(modes)-1)]
    plt.plot(np.array(modes)/1e6, '-o', label = str(n_ions)+' ions')
plt.plot(mode_freq_exp,'--s', label = 'measured')
plt.xlabel('Mode number')
plt.ylabel('Mode frequency (MHz)')
plt.legend()
plt.grid()
```
